```python
import time
import requests
from urllib.request import urlopen as ureq
from bs4 import BeautifulSoup as soup
```

```python
print('Enter Username')
Username=input()
```

# Set the URL 


```python
url="https://github.com/" + Username
url
```

# Connect to the URL

```python
request=ureq(url)#The assembled request
page_html=request.read()
request.close()
```

# Parse HTML 

```python
page_soup=soup(page_html,"html.parser")
```

```python
containers=page_soup.findAll('svg',{'class':"mx-auto js-calendar-graph-svg"})
containers
```

```python
len(containers)
```

```python
soup.prettify(containers[0])
contain=containers[0]
```

```python
days=contain.find_all('rect',{'class':'day'})
days
```

```python
count=[]
date=[]

for i in days:
    count.append(i.get('data-count'))
    date.append(i.get('data-date'))
```

```python
total_days=len(date)
print(total_days)
```

```python
check_days=total_days-45
c=0
for i in range(check_days,total_days):
    c=c+1
print(c)
```

```python
count_push=[]
count_pushh=[]
for i in range(check_days,total_days):
    count_pushh.append(date[i])
    count_push.append(count[i])
print(count_push)
print(count_pushh)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.scatter(count_push,count_pushh,color='magenta')
```
